```python
#import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# Create a scatter plot
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Generate random data
data = np.random.normal(0, 1, 1000)

# Create a histogram
plt.hist(data, bins=50)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Create a line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

These examples demonstrate how we can use Jupyter Notebook, NumPy, and Matplotlib to generate and visualize random data.
