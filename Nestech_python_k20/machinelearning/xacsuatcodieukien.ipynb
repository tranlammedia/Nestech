3. tinh tat ca xs (Ci|X)

X = it, thap, nam
P(X|Mua) = P(it|Mua) x P(Thap|Mua) x P(Nam|Mua) 

        =   0/4   x     1/4     x       1/4
        (vi du lieu it, khong can bang, ap dung laplace correction, Tử luon +1, Mẫu + (số nhãn))
        = (0+1)/(4+2)x(1+1)/(4+3)x(1+1)/(4+2)
        = 1/63

P(X|Khong mua) = P(it|Khong mua) x P(Thap|Khong mua) x P(Nam|Khong mua)

                =   3/4         x       2/4         x       2/4
                = (3+1)/(4+2)   x   (2+1)/(4+3)     x   (2+1)/(4+2)
                = 4/6 x 3/7 x 3/6
                = 2/3 x 3/7 x 1/2
                = 2/7x1/2 = 1/7

P(X) = tổng của (P(X|Ci)xP(Ci))

        = P(X|Mua)xP(Mua) + P(X|Khong mua)xP(Khong mua)

        = 1/63 x 4/8    +       1/7 x 4/8
        (vi du lieu it, khong can bang, ap dung laplace correction)
        = 1/63 x (4+1)/(8+2)    +       1/7 x (4+1)/(8+2)
        = 5/63


P(Mua|X) = P(X|Mua) x P(Mua) / P(X)

        =  1/63     x  4/8   / (5/63)
        = 1/10

P(Khong Mua|X) = P(X|Khong Mua) x P(Khong Mua) / P(X)

                =  1/7     x  4/8   / (5/63)

                = 9/10

X = nhieu, trung binh, nam
P(X|Mua) = P(nhieu|Mua) x P(trung binh|Mua) x P(Nam|Mua) 

        =   4/4   x     1/4     x       1/4
        (vi du lieu it, khong can bang, ap dung laplace correction, Tử luon +1, Mẫu + (số nhãn))
        = (4+1)/(4+2)x(1+1)/(4+3)x(1+1)/(4+2)
        = 5/6 x 2/7 x 2/6
        = 5/63

P(X|Khong mua) = P(nhieu|Khong mua) x P(trung binh|Khong mua) x P(Nam|Khong mua)

                =   1/4         x       0/4         x       2/4
                = (1+1)/(4+2)   x   (0+1)/(4+3)     x   (2+1)/(4+2)
                = 2/6 x 1/7 x 3/6
                = 1/42

P(X) = tổng của (P(X|Ci)xP(Ci))

        = P(X|Mua)xP(Mua) + P(X|Khong mua) x P(Khong mua)

        = 5/63 x 4/8    +       1/42 x 4/8
        (vi du lieu it, khong can bang, ap dung laplace correction)
        = 5/63 x (4+1)/(8+2)    +       1/42 x (4+1)/(8+2)
        = 5/63 x 5/10 + 1/42 x 5/10
        = 13/252


P(Mua|X) = P(X|Mua) x P(Mua) / P(X)

        =  5/63     x  4/8   / (13/252)
        = 10/13

P(Khong Mua|X) = P(X|Khong Mua) x P(Khong Mua) / P(X)

                =  1/42     x  4/8   / (13/252)
                = 3/13
